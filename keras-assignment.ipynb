```python
import pandas as pd
import numpy as np
import requests
# import nessesary library
```

```python
concrete_data = pd.read_csv("https://cocl.us/concrete_data")
concrete_data.head()
```

```python
# make sure that the table is clean
concrete_data.isnull().sum()
```

```python
# separate target and predictors
columns = concrete_data.columns
predictor = concrete_data[columns[columns != "Strength"]]
target = concrete_data["Strength"]
print(predictor.head())
print(target.head())
```

```python
from sklearn.model_selection import train_test_split
```

```python
!pip install keras==2.1.6

import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
num_col = predictor.shape[1]
```

## Task A

```python
def task_a():
    model = Sequential()
    model.add(Dense(10, activation = 'relu', input_shape = (num_col,)))
    model.add(Dense(1))
    
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
import statistics
n = 0
errors = list()
# create a list of 50 mean squared errors
while n < 50:
    X_train, X_test, y_train, y_test = train_test_split(predictor, target,test_size = 0.3, random_state = 6)
    model = task_a()
    model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, verbose=2)
    scores = model.evaluate(X_test, y_test, verbose=0)
    errors.append(scores)
    n += 1

 # Report the mean and the standard deviation of the mean squared errors.
print("mean: {}, standard deviation:{}".format(statistics.mean(errors), statistics.pstdev(errors)))

```

 ## Task B

```python
# normalise
predictor = (predictor -predictor.mean())/predictor.std()
predictor.head()
```

```python
n = 0
errors = list()
# create a list of 50 mean squared errors
while n < 50:
    X_train, X_test, y_train, y_test = train_test_split(predictor, target,test_size = 0.3, random_state = 6)
    model = task_a()
    model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, verbose=2)
    scores = model.evaluate(X_test, y_test, verbose=0)
    errors.append(scores)
    n += 1

```

```python
 # Report the mean and the standard deviation of the mean squared errors.
print("mean: {}, standard deviation:{}".format(statistics.mean(errors), statistics.pstdev(errors)))
```

### Mean is not too different (less than 10% difference) from A but standard deviation is smaller this time.

## Task D

```python
def task_d():
    model = Sequential()
    model.add(Dense(10, activation="relu", input_shape=(num_col,)))
    model.add(Dense(10, activation="relu"))
    model.add(Dense(10, activation="relu"))
    model.add(Dense(1))
    
    model.compile(optimizer = "adam", loss="mean_squared_error")
    return model
    
```

```python
n = 0
errors = list()
# create a list of 50 mean squared errors
while n < 50:
    X_train, X_test, y_train, y_test = train_test_split(predictor, target,test_size = 0.3, random_state = 6)
    model = task_d()
    model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, verbose=2)
    scores = model.evaluate(X_test, y_test, verbose=0)
    errors.append(scores)
    n += 1
```

```python
print("mean: {}, standard deviation:{}".format(statistics.mean(errors), statistics.pstdev(errors)))
```

### Mean and standard deviation are even smaller
